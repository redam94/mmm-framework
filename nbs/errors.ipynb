```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.normal(0, 1, size=(100, 2))
X = np.hstack([X, (X[:, 0] + np.random.normal(0, .02, size=(100,))).reshape(-1, 1)])
y = X @ np.random.normal(0, 1, size=(3,)) + np.random.normal(0, 0.5, size=(100,))
```

```python
inv_XTX = np.linalg.inv(X.T @ X)
beta_hat = inv_XTX @ X.T @ y
y_hat = X @ beta_hat
residuals = y - y_hat

np.random.multivariate_normal(np.zeros(3), inv_XTX/100 @ (X.T @ X * (residuals.T @ residuals)/100) @ inv_XTX/100)
```
